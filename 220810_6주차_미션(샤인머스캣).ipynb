##3-1) 중복되지 않는 학생 3명으로 이루어진 112개의 팀(28개x4)의 팀 구성하기
- 차시가 총 112차시이기 때문

```python
# 각 이름의 열별로 리스트화(append)하고 랜덤(import random)으로 팀 이루기
import csv
name = csv.reader(open('학생이름 3열.csv'))

#팀 구성 작업

n1 = []
n2 = []
n3 = []

# 아래 같은 3개의 명렬표에서 한 명씩 추출하여 총 3명으로 한 팀을 만듦 -> 총 28개의 팀을 이룸

for row in name:
  n1.append(row[0]) #28명 명단 명렬표1
  n2.append(row[1]) #28명 명단 명렬표2
  n3.append(row[2]) #28명 명단 명렬표3 
  
#중복없이 랜덤뽑기 - 학생들이 각 명렬표에서 추출되고 난 뒤에는 다시 뽑히지 않아야 공평한 횟수로 팀원에 분배됨
import random 
import numpy as np
t1 = np.random.choice(n1,28,replace = False)
t2 = np.random.choice(n2,28,replace = False)
t3 = np.random.choice(n3,28,replace = False)

# 팀 구성 - 3명씩 1팀, 총 28팀, 팀 내의 중복X (이게 맞나? 진짜 .. 머리 뽀개지는 줄 알았다.)
g1 = []
g2 = []
g3 = []
g4 = []

#일단 세 명이 한 팀을 이루어 28개의 팀이 들어간 리스트 작성
for i in range(0,28): 
   members = [t1[i],t2[i],t3[i]]
   g1.append(members)
for i in range(0,28): 
   members = [t1[i],t2[i],t3[i]]
   g2.append(members)
for i in range(0,28): 
   members = [t1[i],t2[i],t3[i]]
   g3.append(members)
for i in range(0,28): 
   members = [t1[i],t2[i],t3[i]]
   g4.append(members)   

#28개의 총 팀 중에 중복이 없는 팀을 찾을 때까지 반복
while len(g1) ==28: 
  if t1[i] != t2[i] and t1[i] != t3[i] and t2[i] != t3[i]:
    g1 = g1 
  break
while len(g2) ==28: 
  if t1[i] != t2[i] and t1[i] != t3[i] and t2[i] != t3[i]:
    g2 = g2 
  break
while len(g3) ==28: 
  if t1[i] != t2[i] and t1[i] != t3[i] and t2[i] != t3[i]:
    g3 = g3 
  break
while len(g4) ==28: 
  if t1[i] != t2[i] and t1[i] != t3[i] and t2[i] != t3[i]:
    g4 = g4 
  break

G = g1+g2+g3+g4
print(len(G)) #차시가 총 112개 이므로, 총 팀도 112팀
```
